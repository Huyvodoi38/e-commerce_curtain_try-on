```python
import torch
import cv2
import numpy as np
from PIL import Image, ImageFilter
from segment_anything import sam_model_registry, SamPredictor
from diffusers import StableDiffusionControlNetInpaintPipeline, ControlNetModel, UniPCMultistepScheduler
from transformers import pipeline # Thêm thư viện này cho Depth
import gradio as gr

# --- 1. CẤU HÌNH MỚI ---
# Đổi sang vit_b cho nhẹ
SAM_CHECKPOINT = "sam_vit_b_01ec64.pth"
MODEL_TYPE = "vit_b"

device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32

print(f"🚀 Đang chạy trên thiết bị: {device}")

# --- 2. LOAD MODELS ---

# A. Load SAM (Nhanh hơn)
print("⏳ Đang load SAM (vit_b)...")
sam = sam_model_registry[MODEL_TYPE](checkpoint=SAM_CHECKPOINT)
sam.to(device=device)
sam_predictor = SamPredictor(sam)

# B. Load Depth Estimator (Thay cho Canny)
print("⏳ Đang load Depth Estimator...")
depth_estimator = pipeline('depth-estimation', device=0 if device=="cuda" else -1)

# C. Load ControlNet Depth & Realistic SD
print("⏳ Đang load Stable Diffusion & ControlNet Depth...")

# Dùng ControlNet Depth v1.1 (Chuẩn hơn bản cũ)
controlnet = ControlNetModel.from_pretrained(
    "lllyasviel/control_v11f1p_sd15_depth",
    torch_dtype=dtype
)

# Dùng Model Realistic Vision (Bản chuyên Inpainting)
# Model này cho chất lượng ảnh thật hơn RunwayML gốc
pipe = StableDiffusionControlNetInpaintPipeline.from_pretrained(
    "Uminosachi/realisticVisionV51_v51VAE-inpainting",
    controlnet=controlnet,
    torch_dtype=dtype,
    safety_checker=None
)

pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()

print("✅ --- HOÀN TẤT LOAD MODEL ---")
```

```python
# import gradio as gr
# import numpy as np
# from PIL import Image, ImageFilter
# import cv2
# import torch

# --- 3. CÁC HÀM XỬ LÝ (HELPER FUNCTIONS) ---

def get_depth_map(image):
    """
    Tạo bản đồ chiều sâu (Depth Map) thay vì Canny.
    Giúp AI hiểu không gian 3D của cửa sổ.
    """
    # Depth estimator nhận vào PIL Image
    if isinstance(image, np.ndarray):
        image = Image.fromarray(image)

    # Dự đoán depth
    depth_result = depth_estimator(image)['depth']

    # Depth map trả về là ảnh xám (L), cần convert sang RGB để đưa vào ControlNet
    depth_result = np.array(depth_result)
    depth_result = depth_result[:, :, None]
    depth_result = np.concatenate([depth_result, depth_result, depth_result], axis=2)
    return Image.fromarray(depth_result)

def dilate_mask(mask_array, expansion_amount):
    if expansion_amount <= 0:
        return mask_array
    kernel_size = int(expansion_amount)
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    dilated = cv2.dilate(mask_array, kernel, iterations=1)
    return dilated

def predict_curtain(input_dict, prompt, neg_prompt, expansion, control_scale, process_res):
    # --- BƯỚC 1: LẤY DỮ LIỆU ---
    if input_dict is None or input_dict["background"] is None:
        return None

    image = input_dict["background"]

    # Lấy nét vẽ tay (nếu có)
    user_sketch = None
    if input_dict["layers"] and len(input_dict["layers"]) > 0:
         user_sketch = input_dict["layers"][0]

    image_pil = Image.fromarray(image).convert("RGB")
    w_orig, h_orig = image_pil.size
    process_size = int(process_res)

    print(f"⚙️ Đang xử lý: {process_size}x{process_size}")
    image_resized = image_pil.resize((process_size, process_size), resample=Image.LANCZOS)

    # --- BƯỚC 2: TÍNH BOUNDING BOX ---
    bbox = np.array([0, 0, w_orig, h_orig]) # Mặc định full ảnh

    if user_sketch is not None:
        user_sketch_np = user_sketch[:, :, -1] # Lấy kênh Alpha
        rows = np.any(user_sketch_np, axis=1)
        cols = np.any(user_sketch_np, axis=0)
        if np.any(rows) and np.any(cols):
            y_min, y_max = np.where(rows)[0][[0, -1]]
            x_min, x_max = np.where(cols)[0][[0, -1]]

            # Tính tỉ lệ scale giữa ảnh gốc và layer vẽ
            scale_x = w_orig / user_sketch.shape[1]
            scale_y = h_orig / user_sketch.shape[0]

            bbox = np.array([x_min * scale_x, y_min * scale_y, x_max * scale_x, y_max * scale_y])

    # --- BƯỚC 3: CHẠY SAM (VIT_B) ---
    sam_predictor.set_image(np.array(image_pil))
    masks, _, _ = sam_predictor.predict(point_coords=None, point_labels=None, box=bbox[None, :], multimask_output=False)
    sam_mask_uint8 = (masks[0] * 255).astype(np.uint8)

    # Resize mask
    mask_pil_temp = Image.fromarray(sam_mask_uint8).resize((process_size, process_size), resample=Image.NEAREST)
    mask_np_temp = np.array(mask_pil_temp)

    # --- BƯỚC 4: NỚI RỘNG MASK ---
    scaled_expansion = int(expansion * (process_size / 512))
    mask_dilated_np = dilate_mask(mask_np_temp, scaled_expansion)
    mask_pil = Image.fromarray(mask_dilated_np).convert("L")
    mask_pil = mask_pil.filter(ImageFilter.GaussianBlur(radius=5))

    blurred_image = image_resized.filter(ImageFilter.GaussianBlur(radius=30))

    # 2. Ghép ảnh: Chỉ lấy phần mờ ở chỗ cái Rèm (Mask), còn lại giữ nguyên nét
    # Hàm composite: Nếu mask trắng -> lấy ảnh blur, nếu mask đen -> lấy ảnh gốc
    image_for_inpaint_pil = Image.composite(blurred_image, image_resized, mask_pil)

    # --- BƯỚC 5: GENERATE VỚI DEPTH ---
    # Thay vì Canny, ta gọi hàm get_depth_map
    control_image = get_depth_map(image_resized)

    torch.cuda.empty_cache()

    output = pipe(
        prompt=prompt,
        negative_prompt=neg_prompt,
        image=image_for_inpaint_pil,
        mask_image=mask_pil,
        control_image=control_image,
        num_inference_steps=30,
        guidance_scale=7.5,
        controlnet_conditioning_scale=0.35,
    ).images[0]

    return output.resize((w_orig, h_orig), resample=Image.LANCZOS)

# --- GIAO DIỆN ---
css = """
#img_input {height: 600px !important}
"""

with gr.Blocks(css=css, theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🏠 AI Curtain Try-on (High Quality)")

    with gr.Row():
        with gr.Column():
            input_img = gr.ImageEditor(
                type="numpy",
                label="1. Vẽ bao quanh cửa sổ",
                elem_id="img_input",
                brush=gr.Brush(colors=["#FFFFFF"], color_mode="fixed", default_size=20)
            )

            prompt_text = gr.Textbox(
                label="2. Mô tả rèm",
                value="A pair of white solid curtains, cotton, blackout, highly detailed fabric texture, interior design photography, realistic lighting, 8k, masterpiece, architectural digest style, sharp focus",
                lines=2
            )

            neg_prompt_text = gr.Textbox(
                label="3. Negative Prompt",
                value="window frame distortion, messy, artifacts, bad anatomy, low quality, blurry, broken wall, ugly, text, watermark",
            )

            with gr.Accordion("Cài đặt chất lượng & Rèm", open=True):
                # Slider độ nét (Quan trọng)
                quality_slider = gr.Slider(
                    minimum=512, maximum=1024, value=768, step=128,
                    label="Độ phân giải xử lý (Càng cao càng nét, nhưng chậm)"
                )

                expansion_slider = gr.Slider(
                    minimum=0, maximum=100, value=50, step=5,
                    label="Độ mở rộng rèm (Mask Expansion)"
                )
                control_slider = gr.Slider(
                    minimum=0.1, maximum=1.0, value=0.25, step=0.05,
                    label="Độ bám khung cửa (Thấp = Che kín cửa)"
                )

            run_btn = gr.Button("🚀 Tạo Rèm Sắc Nét", variant="primary")

        with gr.Column():
            output_img = gr.Image(label="Kết quả")

    run_btn.click(
        predict_curtain,
        inputs=[input_img, prompt_text, neg_prompt_text, expansion_slider, control_slider, quality_slider],
        outputs=output_img
    )

demo.launch(share=True, debug=True)
```

```python
# --- 4. BACKEND API (FASTAPI) ĐỂ MOBILE APP GỌI ---

!pip install -q fastapi uvicorn nest_asyncio python-multipart

# ⚠️ QUAN TRỌNG: Import các thư viện cần thiết
import cv2
import numpy as np
from PIL import Image, ImageFilter
import torch

from fastapi import FastAPI, UploadFile, File, Form
from fastapi.middleware.cors import CORSMiddleware
from fastapi.responses import StreamingResponse, JSONResponse
from io import BytesIO

# ⚠️ QUAN TRỌNG: Thêm các helper functions cần thiết
def get_depth_map(image):
    """
    Tạo bản đồ chiều sâu (Depth Map) thay vì Canny.
    Giúp AI hiểu không gian 3D của cửa sổ.
    """
    if isinstance(image, np.ndarray):
        image = Image.fromarray(image)

    depth_result = depth_estimator(image)['depth']
    depth_result = np.array(depth_result)
    depth_result = depth_result[:, :, None]
    depth_result = np.concatenate([depth_result, depth_result, depth_result], axis=2)
    return Image.fromarray(depth_result)

def dilate_mask(mask_array, expansion_amount):
    """Nới rộng mask để rèm che kín hơn"""
    if expansion_amount <= 0:
        return mask_array
    kernel_size = int(expansion_amount)
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    dilated = cv2.dilate(mask_array, kernel, iterations=1)
    return dilated

app = FastAPI(title="AI Curtain Try-on API")

# ⚠️ QUAN TRỌNG: Thêm CORS middleware để frontend có thể gọi API
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # Production: chỉ định domain cụ thể
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Danh sách mẫu rèm (có thể mở rộng sau này hoặc lưu trong DB)
CURTAIN_TEMPLATES = [
    {
        "id": "white_solid",
        "name": "Rèm trắng trơn sang trọng",
        "thumbnail_url": "https://dummyimage.com/600x800/ffffff/000000&text=White+Curtain",
        "prompt": "A pair of white solid curtains, cotton, blackout, highly detailed fabric texture, interior design photography, realistic lighting, 8k, masterpiece, architectural digest style, sharp focus",
        "neg_prompt": "window frame distortion, messy, artifacts, bad anatomy, low quality, blurry, broken wall, ugly, text, watermark",
    },
    {
        "id": "beige_linen",
        "name": "Rèm vải linen màu beige",
        "thumbnail_url": "https://dummyimage.com/600x800/f5f5dc/000000&text=Beige+Linen",
        "prompt": "Beige linen curtains, soft natural fabric, light filtering, modern minimal interior, warm sunlight, realistic lighting, 8k, photography, detailed fabric folds",
        "neg_prompt": "cartoon, illustration, low quality, artifacts, watermark, text",
    },
    {
        "id": "dark_blackout",
        "name": "Rèm tối màu chắn sáng",
        "thumbnail_url": "https://dummyimage.com/600x800/333333/ffffff&text=Dark+Blackout",
        "prompt": "Dark grey blackout curtains, thick fabric, luxurious hotel room style, cinematic lighting, high detail, 8k, interior photography",
        "neg_prompt": "washed out colors, low quality, artifacts, watermark, text",
    },
]


@app.get("/curtains")
async def list_curtains():
    """Trả về danh sách các mẫu rèm cho mobile app hiển thị."""
    return CURTAIN_TEMPLATES


def run_curtain_tryon_core(img_pil, prompt, neg_prompt, bbox, expansion, control_scale, process_res):
    """Core function dùng lại cùng pipeline như trong Gradio, nhưng input là ảnh + bbox."""
    from PIL import ImageFilter

    image_pil = img_pil.convert("RGB")
    w_orig, h_orig = image_pil.size
    process_size = int(process_res)

    # Resize để xử lý
    image_resized = image_pil.resize((process_size, process_size), resample=Image.LANCZOS)

    # --- Tính bounding box ---
    if bbox is None:
        bbox_arr = np.array([0, 0, w_orig, h_orig])
    else:
        x_min, y_min, x_max, y_max = bbox
        # Clamp vào trong ảnh
        x_min = max(0, min(w_orig - 1, x_min))
        x_max = max(1, min(w_orig, x_max))
        y_min = max(0, min(h_orig - 1, y_min))
        y_max = max(1, min(h_orig, y_max))
        bbox_arr = np.array([x_min, y_min, x_max, y_max])

    # --- Chạy SAM để lấy mask vùng cửa ---
    sam_predictor.set_image(np.array(image_pil))
    masks, _, _ = sam_predictor.predict(
        point_coords=None,
        point_labels=None,
        box=bbox_arr[None, :],
        multimask_output=False,
    )
    sam_mask_uint8 = (masks[0] * 255).astype(np.uint8)

    # Resize mask theo kích thước xử lý
    mask_pil_temp = Image.fromarray(sam_mask_uint8).resize(
        (process_size, process_size), resample=Image.NEAREST
    )
    mask_np_temp = np.array(mask_pil_temp)

    # Nới rộng mask
    scaled_expansion = int(expansion * (process_size / 512))
    mask_dilated_np = dilate_mask(mask_np_temp, scaled_expansion)
    mask_pil = Image.fromarray(mask_dilated_np).convert("L")
    mask_pil = mask_pil.filter(ImageFilter.GaussianBlur(radius=5))

    # Làm mờ vùng rèm để inpaint mượt hơn
    blurred_image = image_resized.filter(ImageFilter.GaussianBlur(radius=30))
    image_for_inpaint_pil = Image.composite(blurred_image, image_resized, mask_pil)

    # Depth map cho ControlNet
    control_image = get_depth_map(image_resized)

    torch.cuda.empty_cache()

    output = pipe(
        prompt=prompt,
        negative_prompt=neg_prompt,
        image=image_for_inpaint_pil,
        mask_image=mask_pil,
        control_image=control_image,
        num_inference_steps=30,
        guidance_scale=7.5,
        controlnet_conditioning_scale=control_scale,
    ).images[0]

    return output.resize((w_orig, h_orig), resample=Image.LANCZOS)


@app.post("/tryon")
async def tryon_curtain(
    image: UploadFile = File(..., description="Ảnh chụp cửa sổ"),
    template_id: str = Form(None, description="ID mẫu rèm được chọn"),
    prompt: str = Form(None, description="Prompt tuỳ chỉnh, nếu muốn override template"),
    neg_prompt: str = Form(None, description="Negative prompt tuỳ chỉnh"),
    x_min: int = Form(0, description="Toạ độ x góc trái trên của vùng cửa"),
    y_min: int = Form(0, description="Toạ độ y góc trái trên của vùng cửa"),
    x_max: int = Form(0, description="Toạ độ x góc phải dưới của vùng cửa"),
    y_max: int = Form(0, description="Toạ độ y góc phải dưới của vùng cửa"),
    expansion: int = Form(50, description="Độ nới rộng mask"),
    control_scale: float = Form(0.35, description="Độ bám ControlNet"),
    process_res: int = Form(768, description="Độ phân giải xử lý (512-1024)"),
):
    """Endpoint chính: nhận ảnh + vùng cửa + mẫu rèm, trả về ảnh try-on."""

    # Lấy template nếu có
    tpl = None
    if template_id is not None:
        for t in CURTAIN_TEMPLATES:
            if t["id"] == template_id:
                tpl = t
                break
        if tpl is None:
            return JSONResponse({"error": "Template not found"}, status_code=404)

    final_prompt = prompt or (tpl["prompt"] if tpl else "")
    final_neg_prompt = neg_prompt or (tpl["neg_prompt"] if tpl else "")

    if not final_prompt:
        return JSONResponse({"error": "Prompt is required"}, status_code=400)

    # Đọc ảnh
    img_bytes = await image.read()
    img_pil = Image.open(BytesIO(img_bytes)).convert("RGB")

    # Nếu bbox không hợp lệ thì cho full ảnh
    bbox = None
    if x_max > x_min and y_max > y_min:
        bbox = (x_min, y_min, x_max, y_max)

    try:
        result_pil = run_curtain_tryon_core(
            img_pil,
            final_prompt,
            final_neg_prompt,
            bbox,
            expansion,
            control_scale,
            process_res,
        )
    except Exception as e:
        return JSONResponse({"error": str(e)}, status_code=500)

    buf = BytesIO()
    result_pil.save(buf, format="PNG")
    buf.seek(0)

    return StreamingResponse(buf, media_type="image/png")


# Đoạn này để chạy server trong Colab (có thể chỉnh port tuỳ ý)
import nest_asyncio
import uvicorn
import threading

nest_asyncio.apply()

def run_server():
    """Chạy uvicorn server trong thread riêng để tránh conflict với event loop của Colab"""
    # Tạo event loop mới trong thread này
    import asyncio
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    
    config = uvicorn.Config(app, host="0.0.0.0", port=8000, log_level="info")
    server = uvicorn.Server(config)
    loop.run_until_complete(server.serve())

print("✅ FastAPI server sẵn sàng, truy cập qua ngrok/Colab tunnel:")
print("📡 Server đang chạy tại: http://localhost:8000")
print("📚 API docs: http://localhost:8000/docs")

# Chạy server trong background thread
server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

print("✅ Server đã khởi động! Sử dụng ngrok để expose ra ngoài:")
print("   !pip install pyngrok")
print("   from pyngrok import ngrok")
print("   public_url = ngrok.connect(8000)")
print(f"   print(f'Backend URL: {{public_url}}')")
```
